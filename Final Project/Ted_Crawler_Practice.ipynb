```python
import requests
from bs4 import BeautifulSoup
import re
import urllib.request
```

```python
#考慮下載速度及練習方便性，僅抓六分鐘以下的演講
url_page = "https://www.ted.com/talks?sort=newest&duration=0-6"

# 演講分類
topics = ["Technology", "Entertainment", "Design", "Business", "Science", "Global+issues"]
pick_topic = ""
filtered_url = ""
```

```python
# TODO: 給用戶輸入種類

#目前先以Technology類做爬蟲
pick_topic = topics[0]

if pick_topic == "":
    filtered_url = url_page
else:
    filtered_url = url_page + "&topics%5B%5D=" + pick_topic
print(filtered_url)
```

```python
playlsits_page = requests.get(filtered_url).content
soup = BeautifulSoup(playlsits_page, "html.parser")
print(soup.prettify())
```

```python
#搜尋後可能會有多頁結果，這裏僅將第一頁的所有演講記錄下來
cont = soup.findAll(attrs = {"data-ga-context":"talks"})
cont
```

```python
len(cont)
```

```python
type(cont[0])
```

```python
lists = {}
names = []
for i in range(len(cont)):
    if i%2 != 0:
        uri = cont[i]["href"]
        name = cont[i].text.strip()
        lists[name] = uri
        names.append(name)
```

```python
# lists為字典型別，key為演講主題，value為uri供接下來進入影片頁面使用
lists
```

```python
#到時要show給使用者看的演講主題列表
names
```

```python
# TODO: 給用戶選擇想看的影片

#目前先選擇第一個影片
url_video = "https://www.ted.com" + lists[names[3]]
url_video
```

```python
# 進入影片網頁
video_page = requests.get(url_video)
vedio_cont = video_page.content
vedio_soup = BeautifulSoup(vedio_cont, "html.parser")
print(vedio_soup)
```

```python
# 我們所需資訊在第十個script位置
scripts = vedio_soup.findAll("script")
info_script = scripts[9]
info_script
```

```python
#從text中萃取 (1)audioDownload後方的網址，用來下載音檔 (2)talk_id資訊，用來下載逐字稿

#取得audioDownload_url
info_str = info_script.text
pattern_audio = re.compile('audioDownload":"http.*?mp3.apikey=.*?"')
audio_path = pattern_audio.findall(info_str)
print(audio_path)
split_path = audio_path[0].split("\":\"")
audioDownload_url = split_path[1][:-1]
print(audioDownload_url)
```

```python
#取得id
pattern_id = re.compile('talk_id":\d+')
id_path = pattern_id.findall(info_str)
print(id_path)
split_id = id_path[0].split("\":")
talk_id = split_id[1]
print(talk_id)
transcript_url = "https://www.ted.com/talks/{}/transcript.json?language=en".format(talk_id)
transcript_url
```

```python
#從上面text內容來看
#逐字稿網站位置："https://www.ted.com/talks/23129/transcript.json?language=en"
#音檔下載位置："https://download.ted.com/talks/DeAndreaSalvador_2018U.mp3?apikey=acme-roadrunner"
```

```python
#逐字稿內容
transcript_page = requests.get(transcript_url)
transcript_cont = transcript_page.content
transcript_soup = BeautifulSoup(transcript_cont, "html.parser")
transcript_soup
```

```python
#下載進度
pre=0
def Schedule(a,b,c):
    global pre
    per = 100.0 * a * b / c
    if int(per)-pre>0:
        print('%.2f%%' % per)
        pre=int(per)
```
